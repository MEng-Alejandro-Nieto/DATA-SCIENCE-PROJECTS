## 2. Likes and Dislikes Analysis Per Channel

* Below you will see a data frame indexed by the Youtube Channel name with the sum of likes and dislikes of all their videos and an extra column called ratio which shows how many likes/dislikes the channel has.  This organized in ascending mode by the ratio value 

* You will see a scatter plot labeled by likes and dislikes. One conclusion from the plots is that as the channel becomes more popular, the likes/dislikes ratio decreases drastically. Another finding is that the number of likes can be easily predicted based on the number of interactions with a linear regression model since the number of likes increases proportionally to the number of interactions

```python
likes=data[['channel_title','likes','dislikes']]
likes=likes.groupby('channel_title').sum().sort_values(by='likes',ascending=False)

def func1(x,y):
    if x>0 and y>0:
        return int(x/y)
def func2(x,y):
    if x>0 and y>0:
        return int(x+y)

likes['ratio']=likes[['likes','dislikes']].apply(lambda x: func1(x.likes, x.dislikes), axis=1)
likes['interactions']=likes[['likes','dislikes']].apply(lambda x: func2(x.likes, x.dislikes), axis=1)
likes=likes.sort_values(by='interactions',ascending=False).dropna()
likes.head(5)
```

```python
fig, (axes1,axes2,axes3) =plt.subplots(3,1,figsize=(20,20))

axes1 =sns.scatterplot(data=likes,x='ratio',y='likes',label='LIKES',alpha=0.5,ax=axes1)
axes1 =sns.scatterplot(data=likes,x='ratio',y='dislikes',label='DISLIKES',alpha=0.8,ax=axes1)
axes1.set_xlabel('Ratio')
axes1.set_ylabel('likes-dislikes')
axes1.grid(True)

axes2 =sns.scatterplot(data=likes,x='interactions',y='likes',label='LIKES',alpha=1,ax=axes2)
axes2 =sns.scatterplot(data=likes,x='interactions',y='dislikes',label='DISLIKES',alpha=1,ax=axes2)
axes2.set_xlabel('Interactions')
axes2.set_ylabel('likes-dislikes')
axes2.grid(True)

axes3 =sns.scatterplot(data=likes,x='likes',y='dislikes',alpha=1,ax=axes3)
axes3.set_xlabel('likes')
axes3.set_ylabel('dislikes')
axes3.grid(True)


plt.legend()
```

## 3. Converting Time  string into timestamp

```python
from datetime import datetime
```

```python
data['publish_time']=data['publish_time'].apply(lambda x:datetime.strptime(x.split('T')[0],'%Y-%m-%d'))
data['trending_date']=data['trending_date'].apply(lambda x:'20'+x)
data['trending_date']=data['trending_date'].apply(lambda x:datetime.strptime(x,'%Y.%d.%m'))
```

## 4. Top 10 Channel With More Views and More Comments

```python
data10v=data[['channel_title','views']].groupby('channel_title')['views'].sum().sort_values(ascending=False).head(10)
data10c=data[['channel_title','comment_count']].groupby('channel_title')['comment_count'].sum().sort_values(ascending=False).head(10)
```

```python
fig =plt.figure(figsize=(15,5))
plt.bar(data10v.index,data10v,color='r')
plt.xticks(range(len(data10v)), rotation=45, horizontalalignment='right')
plt.ylabel('Views')
plt.title('Top 10 Channel With More Views')
plt.grid(True)

fig =plt.figure(figsize=(15,5))
plt.bar(data10c.index,data10c,color='b')
plt.xticks(range(len(data10c)), rotation=45, horizontalalignment='right')
plt.ylabel('Comments')
plt.title('Top 10 Channels With More Comments')
plt.grid(True)
```

## 5. Top 10 VEVO Channels Trends in Canada (by Views and Comments)

```python
dataVEVO=data[['channel_title','trending_date','publish_time','views','comment_count','likes','dislikes']].set_index('publish_time').reset_index()
dataVEVO=dataVEVO[dataVEVO['channel_title'].apply(lambda x: True if ('VEVO' in x) else False)]
dataVEVOviews=dataVEVO.groupby('channel_title').sum().sort_values(by='views',ascending=False).head(10)
dataVEVOcomment=dataVEVO.groupby('channel_title').sum().sort_values(by='comment_count',ascending=False).head(10)
```

```python
fig=plt.figure(figsize=(12,6))
plt.bar(dataVEVOviews.index,dataVEVOviews['views']/1000000,color='r')
plt.xticks(range(len(dataVEVOviews)),rotation=45,horizontalalignment='right')
plt.title('Top 10 VEVO videos with more views')
plt.ylabel('Views in millions')
plt.grid(True)
plt.xlabel

fig=plt.figure(figsize=(12,6))
plt.bar(dataVEVOcomment.index,dataVEVOcomment['comment_count']/1000000,color='b')
plt.xticks(range(len(dataVEVOcomment)),rotation=45,horizontalalignment='right')
plt.title('Top 10 VEVO videos with more comments')
plt.ylabel('Comment in millions')
plt.grid(True)
plt.xlabel
```
