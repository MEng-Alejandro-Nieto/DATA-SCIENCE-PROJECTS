# Students in the Departaments

```python
Departamentos=['BOGOTA', 'CUNDINAMARCA', 'ANTIOQUIA', 'VALLE', 'ATLANTICO', 'BOLIVAR', 'NORTE SANTANDER', 'TOLIMA', 'CALDAS', 
'NARIÑO']

MAIN_DEPT=icfes[icfes['ESTU_DEPTO_RESIDE'].apply(lambda x :x in Departamentos)]
g=sns.countplot(MAIN_DEPT['ESTU_DEPTO_RESIDE'],hue=MAIN_DEPT['ESTU_GENERO'])
g.set_xticklabels(g.get_xticklabels(), rotation=40, ha="right")
plt.xlabel('Departamentos')
plt.ylabel('Numero')
plt.title('Numero de Estudiantes - Departamentos')
plt.tight_layout()
plt.grid(True)
```

# Course Average in the Departments

```python
COURSES=['PERCENTIL_MATEMATICAS','PERCENTIL_LECTURA_CRITICA','PERCENTIL_C_NATURALES','PERCENTIL_SOCIALES_CIUDADANAS','PERCENTIL_INGLES','PERCENTIL_GLOBAL']

fem={}
mas={}
for course in COURSES:

    dep_mean_f=[]
    dep_mean_m=[]
    for dep in Departamentos:
        dep_mean_f.append(MAIN_DEPT[(MAIN_DEPT['ESTU_GENERO']=='F') & (MAIN_DEPT['ESTU_DEPTO_RESIDE']==dep)][course].mean())
        dep_mean_m.append(MAIN_DEPT[(MAIN_DEPT['ESTU_GENERO']=='M') & (MAIN_DEPT['ESTU_DEPTO_RESIDE']==dep)][course].mean())
        
    fem.update({course:dep_mean_f})
    mas.update({course:dep_mean_m})


table_mean_f=pd.DataFrame(index=np.array(Departamentos),data=fem,columns=COURSES)
table_mean_m=pd.DataFrame(index=np.array(Departamentos),data=mas,columns=COURSES)
```

# Promedio De Materias Por Departamento

```python
#Diccionario de promedio en matematicas por departamento
dm={}
for dep in Departamentos:
    mean=MAIN_DEPT[MAIN_DEPT['ESTU_DEPTO_RESIDE']==dep]['PERCENTIL_MATEMATICAS'].mean()
    dm.update({dep:mean})
    
def mate(cols,dm=dm):
    return dm[cols[0]]

MAIN_DEPT['Promedio_mate']=MAIN_DEPT[['ESTU_DEPTO_RESIDE']].apply(mate,axis=1)
MAIN_DEPT['Promedio_mate'].unique()
```

# Linear Regression Model
We want to generate a linear regression model to predict the score of the courses with repect to mathematics

```python
from sklearn.cross_validation import train_test_split as tts
from sklearn.linear_model import LinearRegression as lr
```

```python
x=icfes[['PERCENTIL_LECTURA_CRITICA','PERCENTIL_C_NATURALES','PERCENTIL_SOCIALES_CIUDADANAS','PERCENTIL_INGLES','PERCENTIL_GLOBAL']]
y=icfes['PERCENTIL_MATEMATICAS']

x_train, x_test, y_train, y_test = tts(x, y, test_size=0.3, random_state=42)
linear_model=lr()
linear_model.fit(x_train,y_train)
predict=linear_model.predict(x_test)
predict
```

```python
dic_slope={}
dic_inter={}
for course in COURSES:
    x=icfes[[course]]
    y=icfes[['PERCENTIL_MATEMATICAS']]
    x_train, x_test, y_train, y_test = tts(x, y, test_size=0.3, random_state=42)
    linear_model=lr()
    linear_model.fit(x_train,y_train)
    dic_slope.update({course:linear_model.coef_.tolist()})
    dic_inter.update({course:linear_model.intercept_.tolist()})    
    
```

```python
pd.DataFrame(dic_inter)
```

```python
pd.DataFrame(dic_slope)
```

```python
xmat=np.array(icfes['PERCENTIL_MATEMATICAS'])
ycie=xmat*dic_slope['PERCENTIL_C_NATURALES'][0]+dic_inter['PERCENTIL_C_NATURALES'][0]
ysoc=xmat*dic_slope['PERCENTIL_SOCIALES_CIUDADANAS'][0]+dic_inter['PERCENTIL_SOCIALES_CIUDADANAS'][0]
ylec=xmat*dic_slope['PERCENTIL_LECTURA_CRITICA'][0]+dic_inter['PERCENTIL_C_NATURALES'][0]
ying=xmat*dic_slope['PERCENTIL_INGLES'][0]+dic_inter['PERCENTIL_INGLES'][0]
yglo=xmat*dic_slope['PERCENTIL_GLOBAL'][0]+dic_inter['PERCENTIL_GLOBAL'][0]
lista_coef=[[ycie,'Ciencias'],[ysoc,'Sociales'],[ylec,'Lectura'],[ying,'Ingles'],[yglo,'Global']]

fig,ax=plt.subplots(1,1,figsize=(15,8))

for course in lista_coef:
    plt.plot(xmat,course[0],label=course[1])

plt.legend()
plt.grid(True)
```
